🔷 Part A – Theoretical Foundation (ShortNotes & Explanation)




1. What is Statistical Distribution?

 Statistical distribution shows how data values are spread.
 It tells us which values are common and which are rare.

 - Example: Most transactions may be between ₹100–₹500.
 - Simple explain line: “Statistical distribution helps us understand how data is spread.”

2. What is Q-Q Plot and why is it used?


 Q-Q Plot is used to check if data follows a normal distribution.
 - If points are in a straight line → data is normal
 - If points are scattered → not normal

  In a Simple way “Q-Q plot helps us check whether data is normally distributed or not.”

3. Difference between Discrete and Continuous Distribution

- Discrete Data:

Countable values

Example: number of transactions

- Continuous Data:

Can take any value (decimal)

Example: transaction amount

In a Simple way “Discrete data is countable, while continuous data can take any value.”

4. What is Bernoulli Distribution?

 Bernoulli distribution has only two outcomes:

- Success (1)

- Failure (0)

 - Example: transaction success or fail

 - Simple line  “Bernoulli distribution is used when there are only two possible outcomes.”

5. What is Binomial Distribution?

 Binomial distribution counts number of successes in multiple trials.

 - Example:
 Out of 10 transactions, how many are successful?

 - Simple line:
 “Binomial distribution counts successes in repeated experiments.”

6. Explain Log-Normal Distribution

 Log-normal distribution means:
 If we take log of data, it becomes normal.

 - Mostly used for money data
 - Data is right-skewed

  Simple line:
 “Log-normal distribution is used for skewed data like transaction amounts.”

7. Explain Power Law Distribution

 Power law means:
  Few values are very large
   Most values are small

 - Example: few customers spend very high money

  Simple line:
 “Power law shows that a small number of values are very large.”

8. What is Box-Cox Transform?

 Box-Cox transform is used to make data normal.

 - It reduces skewness
 - Helps in better analysis

Simple line: “Box-Cox transform helps to normalize skewed data.”

9. Explain Poisson Distribution with Example

 Poisson distribution is used to count events in a fixed time.

 - Example: number of transactions per day

   Poisson distribution Stands For models how many times an event happens in a fixed interval

10. What is Z-score Probability?

 Z-score shows how far a value is from the mean.

 - Formula:
  = (value - mean) / standard deviation

  Simple line: “Z-score tells how far a value is from the average.”

11. Difference between PDF and CDF

  - PDF (Probability Density Function):

 Shows probability of exact value

 Used for continuous data

  - CDF (Cumulative Distribution Function):

 Shows probability up to a value

 Always increasing

  In Simple way “PDF shows 'probability density', while CDF shows 'cumulative probability'.”
